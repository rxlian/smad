```python
url = 'https://www.adl.org/hate-symbols'
path = 'images/'
count = 0
def fetch_image(count, url):
    request = urllib.request.Request(url)
    response = urllib.request.urlopen(request)
    buf = response.read()
    buf = str(buf, encoding='utf-8')
    listurl = re.findall(r'http.+\.jpg', buf)
    for url in listurl:
        f = open(path+str(count)+'.jpg', 'wb+')
        req = urllib.request.urlopen(url)
        buf = req.read()
        # buf = str(buf)
        f.write(buf)
        count += 1
    return count
```

```python
def get_hyper_links(wd):
    urls, image_urls = [], []
    hplink = wd.find_elements_by_xpath("//div/div[3]/a")
    images = wd.find_elements_by_xpath("//div/div[3]/a/img")
    for element in hplink:
        link = element.get_attribute('href')
        urls.append(link)
        urls[:] = list(OrderedDict.fromkeys(urls))
    for image in images:
        image_link = image.get_attribute('src')
        image_urls.append(image_link)
    return urls, image_urls
```

```python
# path = 'images/'
url = 'https://www.adl.org/hate-symbols'
wd.get(url)
all_url, all_image_url = [], []
count = 0

while True:
    print(f"page:{count}")
    urls, image_urls = get_hyper_links(wd)
    print(len(urls))
    print(len(image_urls))
    all_url.append(urls)
    all_image_url.append(image_urls)

    next_page = wd.find_elements_by_xpath("//a[@class='next pager uppercase icon-textlink']")
    if len(next_page) != 0:
        next_page_link = next_page[0].get_attribute('href')
        wd.get(next_page_link)
        count += 1
        time.sleep(5)
    else:
        print("last page")
        break
# all_url = [item for sublist in all_url for item in sublist]
```
